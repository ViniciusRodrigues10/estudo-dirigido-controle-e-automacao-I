```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, step_response, bode, nyquist
from control import tf, bode_plot
plt.rcParams['figure.figsize'] = (8,5)
```

```python
K = 70
a = 0.6
num = [K*a]
den = [1, 8, 17, (K+10), a*K]
G = tf(num, den)

t, y = step_response(G)
plt.plot(t, y)
plt.title("Veículo com Esteiras — Resposta ao Degrau")
plt.xlabel("Tempo (s)")
plt.ylabel("Saída")
plt.grid(True)
plt.show()
```

```python
Kp = 20
Kd = 3
a1, a2, a3 = 9.1, 2.7, 1.35
tau = 0.2

num = [Kp*a2 - a1]
den = [tau, (1 + Kd*a3), (Kd*a2 + Kp*a3 - tau*a1), Kp*a2 - a1]
G = tf(num, den)

t, y = step_response(G)
plt.plot(t, y)
plt.title("Motocicleta Robótica — Resposta ao Degrau")
plt.xlabel("Tempo (s)")
plt.ylabel("Ângulo (rad)")
plt.grid(True)
plt.show()
```

```python
# Função de transferência do disco rígido
G = tf([1], [1e-6, 0.02, 1])

# Apenas plota (não tenta armazenar retorno)
bode_plot(G, dB=True, Hz=False, deg=True, omega_limits=[1e1, 1e5], omega_num=500)

plt.suptitle("Disco Rígido — Resposta em Frequência (Bode)", y=1.02)
plt.show()
```

```python
from control import margin
G = tf([10], [1, 4, 13, 10])
gm, pm, wg, wp = margin(G)
print("Margem de ganho:", gm, "\nMargem de fase:", pm)
nyquist(G)
plt.title("Diagrama de Nyquist — Margem de Estabilidade")
plt.grid(True)
plt.show()
```

## Conclusão
Os casos apresentados mostram como o conceito de estabilidade é aplicado em contextos industriais:
- O **veículo com esteiras** e a **motocicleta robô** ilustram estabilidade dinâmica real.
- O **disco rígido** mostra como o controle evita vibrações destrutivas.
- O **Nyquist** e **margens de fase/ganho** permitem garantir robustez.

Esses conceitos são fundamentais em **robótica, APC, DCS e Digital Twins** da Indústria 4.0.
